```python
import pandas as pd

file_path = 'train.csv'
data = pd.read_csv(file_path)

summary = {
    "shape": data.shape,
    "columns": data.columns.tolist(),
    "missing_values": data.isnull().sum().to_dict(),
    "sample_data": data.head().to_dict(orient="records")
}

summary
```

```python
# Perform exploratory data analysis (EDA)

# Count of unique labels
unique_labels = data['label'].nunique()
label_counts = data['label'].value_counts()

# Sentence length analysis
data['sentence_length'] = data['sentence'].apply(len)

# Compile EDA results
eda_results = {
    "unique_labels": unique_labels,
    "label_distribution": label_counts.to_dict(),
}

eda_results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing label distribution
plt.figure(figsize=(10, 6))
sns.barplot(y=label_counts.index, x=label_counts.values, palette="viridis")
plt.title("Label Distribution", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Labels", fontsize=12)
plt.tight_layout()
plt.show()

# Sentence length distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['sentence_length'], kde=True, bins=30, color="skyblue")
plt.title("Sentence Length Distribution", fontsize=16)
plt.xlabel("Sentence Length", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()

# Boxplot for sentence lengths grouped by labels
plt.figure(figsize=(12, 8))
sns.boxplot(y='label', x='sentence_length', data=data, palette="coolwarm")
plt.title("Sentence Length by Label", fontsize=16)
plt.xlabel("Sentence Length", fontsize=12)
plt.ylabel("Labels", fontsize=12)
plt.tight_layout()
plt.show()
```
